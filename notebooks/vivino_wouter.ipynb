```python
connection = sqlite3.connect(db_path)
cursor = connection.cursor()


""""We want to highlight 10 wines to increase our sales. Which ones should we choose and why?
- Those with good ratings but not a big amount of ratings
- Natural wines?"""


cursor.execute('''SELECT w.name,w.ratings_average,w.ratings_count FROM wines w
                WHERE w.is_natural = 1
                ORDER by ratings_count,w.ratings_average DESC  LIMIT 10''')


rows = cursor.fetchall()
for row in rows:
    print(row)

connection.close()
```

```python
"""
We have detected that a big cluster of customers like a specific combination of tastes.
We have identified a few primary keywords that match this.
We would like you to find all the wines that have those keywords. 
To ensure the accuracy of our selection, ensure that more than 10 users confirmed those keywords.
 Also, identify the group_name related to those keywords. 
 - coffee
- toast
- green apple
- cream
- citrus
"""
connection = sqlite3.connect(db_path)
cursor = connection.cursor()

cursor.execute(''' SELECT w.name,k.name,kw.count,kw.group_name FROM wines AS w
                   JOIN keywords_wine kw ON kw.wine_id = w.id
                   JOIN keywords k ON kw.keyword_id = k.id 
                   WHERE (k.name = "coffee" OR k.name = "toast" OR k.name ="green apple" OR k.name="cream" OR k.name="citrus")AND kw.count > 10
                   ORDER BY kw.count DESC
                ''')

rows = cursor.fetchall()

for row in rows:
    print(row)

connection.close()

```

```python


"""
We would like to select wines that are easy to find all over the world.
"""
connection = sqlite3.connect(db_path)
cursor = connection.cursor()
cursor = connection.cursor()
cursor.execute('''SELECT w.name, COUNT(c.name) AS num_countries
                FROM wines AS w
                JOIN regions AS r ON w.region_id = r.id
                JOIN countries AS c ON r.country_code = c.code
                GROUP BY w.name
                ORDER BY num_countries DESC;''')
rows = cursor.fetchall()

for row in rows:
    print(row)

connection.close()
```

```python
connection = sqlite3.connect(db_path)
cursor = connection.cursor()

"""Find the top 3 most common grape all over the world  """
print("Top 3 most common grape")
cursor.execute('''SELECT mug.grape_id,g.name,COUNT(grape_id) AS grape_count FROM most_used_grapes_per_country AS mug
               JOIN grapes AS g ON mug.grape_id = g.id
               GROUP BY grape_id
               ORDER BY grape_count DESC
               LIMIT 3''')   
               
rows = cursor.fetchall()

for row in rows:
    print(row)
print("\nTop 5 each grape")

"""and for each grape, give us the the 5 best rated wines."""
cursor.execute('''SELECT sub.ratings_average,sub.name, sub.grape_id
                FROM (
                    SELECT w.ratings_average, w.name, mug.grape_id, ROW_NUMBER() OVER(PARTITION BY mug.grape_id ORDER BY w.ratings_average DESC) AS rank
                    FROM wines AS w
                    JOIN regions r ON w.region_id = r.id
                    JOIN countries c ON r.country_code = c.code
                    JOIN most_used_grapes_per_country AS mug ON mug.country_code = c.code
                    WHERE mug.grape_id = 2 OR mug.grape_id = 10 OR mug.grape_id = 5
                ) AS sub
                WHERE sub.rank <= 5
                ORDER BY sub.grape_id, sub.rank;
                ''')
rows = cursor.fetchall()
y=0
for x in rows:
        print(x)
    

connection.close()
```

```python
""""We would to give create a country leaderboard,
give us a visual that shows the average wine rating for each country."""

connection = sqlite3.connect(db_path)
cursor = connection.cursor()

cursor.execute("""SELECT ROUND(AVG(w.ratings_average),2) AS average,c.name FROM wines AS w
               JOIN regions AS r ON w.region_id = r.id
               JOIN countries AS c ON r.country_code = c.code
               GROUP BY c.code
               ORDER BY average DESC""")
rows = cursor.fetchall()

for row in rows:
    print(row)
print("\nVintages")

""""Do the same for the vintages."""
cursor.execute("""SELECT c.name, ROUND(AVG(v.ratings_average),2) AS average FROM wines AS w
               JOIN vintages AS v ON v.wine_id = w.id  
               JOIN regions AS r ON w.region_id = r.id
               JOIN countries AS c ON r.country_code = c.code
               GROUP BY c.code
               ORDER BY average DESC """)

rows = cursor.fetchall()

for row in rows:
    print(row)


connection.close()
```
