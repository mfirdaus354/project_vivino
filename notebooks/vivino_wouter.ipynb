```python
import sqlite3
db_path = r'data\vivino.db'
connection = sqlite3.connect(db_path)
cursor = connection.cursor()

"""
We would like to select wines that are easy to find all over the world.
"""
cursor.execute('''SELECT w.name, COUNT(c.name) AS num_countries
                FROM wines AS w
                JOIN regions AS r ON w.region_id = r.id
                JOIN countries AS c ON r.country_code = c.code
                GROUP BY w.name
                ORDER BY num_countries DESC;''')
rows = cursor.fetchall()

for row in rows:
    print(row)

connection.close()
```
